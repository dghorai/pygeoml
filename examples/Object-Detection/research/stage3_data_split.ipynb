```python
def create_directory(path_to_directory: list, verbose=True):
    for path in path_to_directory:
        os.makedirs(path, exist_ok=True)
        if verbose:
            print(f"create directory at: {path}")
```

```python
import glob
import xml.etree.ElementTree as ET
import pandas as pd

def xml_to_csv(path):
    xml_list = []
    for xml_file in glob.glob(path + '/*.xml'):
        tree = ET.parse(xml_file)
        root = tree.getroot()
        for member in root.findall('object'):
            value = (root.find('filename').text,
                     int(root.find('size')[0].text),
                     int(root.find('size')[1].text),
                     member[0].text,
                     int(member[4][0].text),
                     int(member[4][1].text),
                     int(member[4][2].text),
                     int(member[4][3].text)
                     )
            xml_list.append(value)
    column_name = ['filename', 'width', 'height',
                   'class', 'xmin', 'ymin', 'xmax', 'ymax']
    xml_df = pd.DataFrame(xml_list, columns=column_name)
    return xml_df
```

```python
IMG_ROOT = "artifacts/data/images/"
ANNOTATION_ROOT = "artifacts/data/annotations/xmls/"

xml_df = xml_to_csv(ANNOTATION_ROOT)
xml_df
```

```python
from sklearn.model_selection import train_test_split

RANDOM_STATE = 42
TRAIN_SIZE, VAL_SIZE, TEST_SIZE = 0.7, 0.1, 0.2


train_valid, test = train_test_split(xml_df, test_size=TEST_SIZE, random_state=RANDOM_STATE, shuffle=True)

train, valid = train_test_split(train_valid, test_size=VAL_SIZE, random_state=RANDOM_STATE, shuffle=True)
```

```python
print(f'Training Data: {train.shape}')
print(f'\nValidation Data: {valid.shape}')
print(f'\nTesting Data: {test.shape}')
```

```python
import shutil

from pathlib import Path

# save data into seperate folder
def copy_files(df, dst, index='filename'):
    for i, row in df.iterrows():
        file_name = row[index]
        src_file = Path(IMG_ROOT, file_name)
        shutil.copy2(src_file, dst)
    return

files = sorted([os.path.join(IMG_ROOT, fname) for fname in os.listdir(IMG_ROOT) if fname.endswith(".jpg")])

# create directory
train_dir = 'research/data/data_split/train'
valid_dir = 'research/data/data_split/valid'
test_dir = 'research/data/data_split/test'
create_directory([train_dir, valid_dir, test_dir])

copy_files(train, train_dir, index='filename')
copy_files(valid, valid_dir, index='filename')
copy_files(test, test_dir, index='filename')
```
